## BQM for the Backpacker Traveller  Problem

### Formulate BQM for the Backpacker Traveller  Problem by using the functions provided by Ocean SDK

#### Problem Statement: 
Given a set of temples and corresponding distances between each pair of temples, the goal is to find the shortest possible route such that a backpacker visits every temple exactly once and returns to the starting point which is known as Hamiltanion Cycle

#### Importing necessary tools and packages

```python
import matplotlib.pyplot as plt
import numpy as np

import dimod
from dimod import BQM
import dwave_networkx as dnx
import networkx as nx
from neal import SimulatedAnnealingSampler
```

#### Created a visualization function to picturize the data

```python
def graph_viz(G):
    """Visualize NetworkX graph.
    """
    options = {
        "node_size": 900,
        "width": 5,
        "arrowstyle": "->",
        "arrowsize": 18,
        "font_size": 20,
        "font_weight": "bold",
        "font_color": "yellow",
    }
    pos = nx.kamada_kawai_layout(G)  # pos = nx.nx_agraph.graphviz_layout(G)
    nx.draw_networkx(G, pos, **options)
    labels = nx.get_edge_attributes(G, "weight")
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.axis("off")
    plt.show()
```

### Vertices Representation and their distance(Km):
#### 0 - Rameshwaram Temple, 
#### 1 - Shree Dwarkadhish Temple,
#### 2 - Kedarnath Temple,
#### 3 - Konark Sun Temple,

```python

diagram = nx.Graph()
diagram.add_weighted_edges_from({(0, 1, 2276), (0, 2, 2887), (0, 3, 1790), (1, 2, 1659), (1, 3, 2038), (2, 3, 1788)})
graph_viz(diagram)
```

#### Using simulated annealing sampler we will optimize the path in a way that a backpacker should take to cover all the temples in short distance

```python
sampler = SimulatedAnnealingSampler()
sampler
```

```python
BTP = dnx.traveling_salesperson(diagram, sampler, start=0) # backpacker is starting at rameswaram temple
print(BTP)
```

## [0, 3, 1, 2] this is the  optimal route backpacker has to take
